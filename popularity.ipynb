```python
def get_movies(genre, language, n=10):
    code = language_map[language.lower()]
    df=movie[(movie['genres'].apply(lambda x: genre in x)) & (movie['original_language'] == code)].copy()
    if df.empty:
        return "Not available"
    C = df['vote_average'].mean()
    m = df['vote_count'].quantile(0.90)
    qualified = df[df['vote_count'] >= m].copy()
    if qualified.empty:
        qualified = df.copy()
    qualified['score'] = qualified.apply(lambda x: (x['vote_count']/(x['vote_count']+m)*x['vote_average'])+(m/(x['vote_count']+m)*C),axis=1)
    return qualified.sort_values('score', ascending=False)[['title','genres','vote_average','vote_count','score']].head(n)
```
